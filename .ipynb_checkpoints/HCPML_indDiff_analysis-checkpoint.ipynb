## Import packages
This pipeline primarly uses pymvpa (to install: http://www.pymvpa.org/installation.html), and also wraps some algorithms in scikit-learn.  

```python
from mvpa2.suite import *
from mvpa2.clfs.skl.base import SKLLearnerAdapter
import os
import pylab
import random
import numpy as np
import nibabel as nib
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
import seaborn as sns
%matplotlib inline  
```

## Define paths
Replace paths with machine-specific paths. 

```python
#enable output to console
verbose.level = 2

#define paths
task      = 'WM' #motor, WM, gambling
data_path = os.path.join('/Volumes/maloneHD/Data/HCP_ML/', task)
beta_path = os.path.join('/Volumes/maloneHD/Data_noSync/HCP_ML/', task)  
mvpa_path = os.path.join(data_path,'mvpa')
```

## Set analysis parameters 
nsubs: list of sample sizes to run 
<br>
targets: str corresponding to the parameter estimates (aka beta images). you can name this whatever you want 

```python
#analysis parameters
nsubs    =  [10,50,100,250,500,1000] #sample sizes to be run
model    = 'randForest'#SVR, SVRrbf, lasso, ridge, gpr, randForest
cv_type  = 'LOSO' #LOSO, nfld (n-fold)
if task is 'WM':
    targets  = ['2BKgt0BK'] #name for parameter estimate
    pe_num   = ['11'] #parameter estimate, aka beta estimate
elif task is 'gambling':
    targets  = ['reward']
    pe_num   = ['2']

#define subjects and mask
subs     = os.listdir(beta_path)
subs     = subs[:max(nsubs)]
```

## Load behavioral data 
Load behavioral data into pandas dataframe, and grab the behavioral data for the behaiovral metric of interest. 

```python
#load behavioral data
df      = pd.read_csv('data/behav_data.csv')
isubs   = [int(s) for s in subs] #convert str to int
df2     = df.loc[df['Subject'].isin(subs)]
if task is 'WM':
    bdata = df2.ListSort_AgeAdj
elif task is 'gambling':
    bdata = df2.DDisc_AUC_200
bdata = bdata.values 
```

## Load beta images
Load beta images into a pymvpa dataset, add the behavioral metric to be predicted as the 'targets', set the subject atributes ('sa') to 1:nsubs, and finally stack all the datasets in 'fds'.
 

```python
#load beta imgs
ds_all = []
for index, s in enumerate(subs):
    tds_beta_path = os.path.join(beta_path, s,
                                 'MNINonLinear', 'Results', 'tfMRI_'+task,
                                 'tfMRI_'+task+'_hp200_s2_level2.feat',
                                 'GrayordinatesStats')
    pe_paths = []
    for p in pe_num:
        pe_path = os.path.join(tds_beta_path,
                                     'cope'+p+'.feat','pe1.dtseries.nii')
    if math.isnan(bdata[index]): #if behavioral metric is missing, replace with sample mean
        bdata[index] = np.nanmean(bdata)
    ds = fmri_dataset(pe_path, targets=bdata[index])
    ds.sa['subject'] = [int(s)]
    ds_all.append(ds)
    #verbose(2, "subject %i of %i loaded" % (index+1, max(nsubs)))

fds = vstack(ds_all) #stack datasets
```

## Set model type, cross-validation scheme, and run cross-validation 
The NFoldPartitioner is a little confusing (documentation here: http://www.pymvpa.org/generated/mvpa2.generators.partition.NFoldPartitioner.html). The 'cvtype' argument is a float between 0 and 1 that specifies the ratio of samples to be left out in the validation test, e.g. 0.2 means 20% of the subjects will be left out for testing. The 'count' variable is the number of splits, or CV folds, e.g. count=5 runs a 5-fold CV. 
<br>
'errorfx' should be set to 'None' in order to save out the raw predicted values. 

```python
#classifier algorithm
if model is 'SVR':
   # clf = LinearCSVMC(tube_epsilon=0.01)
    from sklearn.svm import LinearSVR
    clf = SKLLearnerAdapter(LinearSVR())
elif model is 'SVRrbf':
    from sklearn.svm import SVR
    clf = SKLLearnerAdapter(SVR(kernel='rbf',C=1e03))
elif model is 'lasso':
    from sklearn.linear_model import Lasso
    clf = SKLLearnerAdapter(Lasso())
elif model is 'decisionTree':
    from sklearn.tree import DecisionTreeRegressor
    clf = SKLearnerAdapter(DecisionTreeRegressor())
elif model is 'ridge':
    from sklearn.linear_model import RidgeCV
    clf = SKLLearnerAdapter(RidgeCV())
elif model is 'randForest':
    from sklearn.ensemble import RandomForestRegressor
    clf = SKLLearnerAdapter(RandomForestRegressor(n_estimators=10,max_features='sqrt'))
elif model is 'gpr':
    from sklearn.gaussian_process import GaussianProcessRegressor
    clf = SKLLearnerAdapter(GaussianProcessRegressor(normalize_y=True))
elif model is 'kernelRidge':
    from sklearn.kernel_ridge import KernelRidge
    clf = SKLLearnerAdapter(KernelRidge(kernel='linear'))
        
#cross-validation algorithm
if cv_type is 'LOSO': #leave-one-subject out 
    cv = CrossValidation(clf,
                         NFoldPartitioner(attr='subject'),
                         errorfx=None,
                         pass_attr='subject')
elif cv_type is 'nfld':
    cv = CrossValidation(clf,
                         NFoldPartitioner(cvtype=0.2,count=5,
                                         selection_strategy='random', attr='subject'),
                         errorfx=None,
                         enable_ca=['stats'],
                         pass_attr='subject')

    
#run classification for all sample sizes
# test = random.sample(range(50),10)
# fds[test]
cv_out = []
for n in nsubs[5:6]:
    rndsubs = random.sample(range(1000),n)
    tmp = cv(fds[rndsubs])
    cv_out.append(tmp)
    verbose(2, "sample size of %i subjects analyzed" % n)
```

## Create pandas dataframe and save results 


```python
df = pd.DataFrame([])
for index, n in enumerate(nsubs[5:6]):
    t    = cv_out[index].targets.reshape(n,1) #targets
    p    = cv_out[index].samples #predictions
    pe   = np.abs(t-p)
    s    = cv_out[index].sa.subject #subject #
    sdf  = pd.DataFrame(data=s,columns=['Subject # n=' +str(n)])
    tdf  = pd.DataFrame(data=t,columns=['Targets n=' +str(n)])
    pdf  = pd.DataFrame(data=p,columns=['Prediction n=' +str(n)])
    pedf = pd.DataFrame(data=pe,columns=['Prediction error n=' +str(n)])
    df   =  pd.concat([df,sdf,tdf,pdf,pedf],axis=1)
    
df.to_csv(os.path.join(mvpa_path,task+'_'+model+'_indDiffPredict.csv'))
```
